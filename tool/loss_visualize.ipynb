```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Load the data
# df = pd.read_csv("output_losses.csv")

# Get unique epochs to plot each separately
unique_epochs = df["Epoch"].unique()

# Create subplots with one row for each epoch
fig = make_subplots(
    rows=len(unique_epochs), 
    cols=1,
    shared_xaxes=True,
    subplot_titles=[f"Epoch {epoch}" for epoch in unique_epochs]
)

for i, epoch in enumerate(unique_epochs, start=1):
    epoch_data = df[df["Epoch"] == epoch]
    
    # Add Classification Loss trace
    fig.add_trace(go.Scatter(
        x=epoch_data["Iteration"],
        y=epoch_data["Loss Action CLS"],
        mode='lines+markers',
        name=f'Epoch {epoch} - Classification Loss',
        hovertemplate='Iteration: %{x}<br>Classification Loss: %{y}'
    ), row=i, col=1)
    
    # Add Pose Gate Loss trace
    fig.add_trace(go.Scatter(
        x=epoch_data["Iteration"],
        y=epoch_data["Loss Pose Gate"],
        mode='lines+markers',
        name=f'Epoch {epoch} - Pose Gate Loss',
        hovertemplate='Iteration: %{x}<br>Pose Gate Loss: %{y}'
    ), row=i, col=1)

# Update layout for the entire figure
fig.update_layout(
    height=300 * len(unique_epochs),  # Adjust height based on the number of epochs
    title="Losses vs Iterations Across Epochs",
    xaxis_title="Iteration",
    yaxis_title="Loss",
    hovermode="x unified",
    showlegend=True,
    template="plotly_dark"  # Optional: Choose a theme
)

# Save as an HTML file
fig.write_html("all_epochs_loss_plot.html")
```

```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Load the data
# df = pd.read_csv("output_losses.csv")

# Get unique epochs to plot each separately
unique_epochs = df["Epoch"].unique()

# Create subplots with one row for each epoch
fig = make_subplots(
    rows=len(unique_epochs), 
    cols=1,
    shared_xaxes=True,
    subplot_titles=[f"Epoch {epoch}" for epoch in unique_epochs]
)

for i, epoch in enumerate(unique_epochs, start=1):
    epoch_data = df[df["Epoch"] == epoch]
    
    # Calculate average losses for the epoch
    avg_loss_action_cls = epoch_data["Loss Action CLS"].mean()
    avg_loss_pose_gate = epoch_data["Loss Pose Gate"].mean()
    
    # Add Classification Loss trace
    fig.add_trace(go.Scatter(
        x=epoch_data["Iteration"],
        y=epoch_data["Loss Action CLS"],
        mode='lines+markers',
        name=f'Epoch {epoch} - Classification Loss',
        hovertemplate='Iteration: %{x}<br>Classification Loss: %{y}'
    ), row=i, col=1)
    
    # Add Pose Gate Loss trace
    fig.add_trace(go.Scatter(
        x=epoch_data["Iteration"],
        y=epoch_data["Loss Pose Gate"],
        mode='lines+markers',
        name=f'Epoch {epoch} - Pose Gate Loss',
        hovertemplate='Iteration: %{x}<br>Pose Gate Loss: %{y}'
    ), row=i, col=1)
    
    # Add horizontal lines for average losses
    fig.add_shape(
        type="line",
        x0=epoch_data["Iteration"].min(),
        x1=epoch_data["Iteration"].max(),
        y0=avg_loss_action_cls,
        y1=avg_loss_action_cls,
        line=dict(dash="dash", color="blue"),
        row=i, col=1
    )
    fig.add_annotation(
        x=epoch_data["Iteration"].max(),
        y=avg_loss_action_cls,
        text=f"Avg Classification Loss: {avg_loss_action_cls:.2f}",
        showarrow=False,
        xanchor="left",
        row=i, col=1
    )

    fig.add_shape(
        type="line",
        x0=epoch_data["Iteration"].min(),
        x1=epoch_data["Iteration"].max(),
        y0=avg_loss_pose_gate,
        y1=avg_loss_pose_gate,
        line=dict(dash="dash", color="orange"),
        row=i, col=1
    )
    fig.add_annotation(
        x=epoch_data["Iteration"].max(),
        y=avg_loss_pose_gate,
        text=f"Avg Pose Gate Loss: {avg_loss_pose_gate:.2f}",
        showarrow=False,
        xanchor="left",
        row=i, col=1
    )

# Update layout for the entire figure
fig.update_layout(
    height=300 * len(unique_epochs),  # Adjust height based on the number of epochs
    title="Losses vs Iterations Across Epochs with Average Losses",
    xaxis_title="Iteration",
    yaxis_title="Loss",
    hovermode="x unified",
    showlegend=True,
    template="plotly_dark"  # Optional: Choose a theme
)

# Save as an HTML file
fig.write_html("all_epochs_loss_plot_with_averages.html")
```
